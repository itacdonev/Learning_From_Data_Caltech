### Question 3
Which of the following statements is *false*?

(b) If there is overfitting, there must be two or more hypotheses that have different values of $E_{out}$.

(d) We can always determine if there is overfitting by comparing the values of $(E_{out} - E_{in})$.  
(e) We cannot determine overfitting based on one hypothesis only.  

**Answer**: From Mitchell (p.67) we have the following definition
> Given a hypothesis space $\mathcal H$, a hypothesis $h \in \mathcal H$ is said to overfit the training data if there exists some alternative hypothesis $h' \in \mathcal H$, s.t. $h$ has smaller error than $h'$ over the training examples, but $h'$ has a smaller error than $h$ over the entire distribution of instances.  

Given the definition we can write the following $E^{(h)}_{in} < E^{(h')}_{in}$  and $E^{(h)}_{out} > E^{(h')}_{out}$. Also examples from the book on p.121 can be applied.

Let's check each statement separately:  
(a) If there is overfitting, there must be two or more hypotheses that have different values of $E_{in}$. $\to$ <span style="color:green">True</span> Result from above definition  

(b) If there is overfitting, there must be two or more hypotheses that have different values of $E_{out}$.  $\to$ <span style="color:green">True</span> Result from above definition  

(c) If there is overfitting, there must be two or more hypotheses that have different values of $(E_{out} - E_{in})$. $\to$ <span style="color:green">True</span> Result from above definition    

(d) We can always determine if there is overfitting by comparing the values of $(E_{out} - E_{in})$.  $\to$ <span style="color:salmon">False</span> Can not base the decision on one hypothesis only as per definition.  

(e) We cannot determine overfitting based on one hypothesis only. $\to$ <span style="color:green">True</span> Given the definition above we need two hypothesis.   

<span style="color:blue">Answer is (d).</span>

## Regularized Linear Regression
We are going to experiment with linear regression for classification on the processed US Postal Service Zip Code data set from Homework 8. Download the data (extracted features of intensity and symmetry) for training and testing:   
http://www.amlbook.com/data/zip/features.train  
http://www.amlbook.com/data/zip/features.test  
(the format of each row is: digit intensity symmetry). We will train two types of binary classifiers; one-versus-one (one digit is class +1 and another digit is class -1, with the rest of the digits disregarded), and one-versus-all (one digit is class +1 and the rest of the digits are class -1). When evaluating $E_{in}$ and $E_{out}$, use binary classification error. Implement the regularized least-squares linear regression for classification that minimizes
$$
\frac{1}{N} \sum^{N}_{n=1} (\mathbf w^Tz_n-y_n)^2 + \frac{\lambda}{N}\mathbf w^T\mathbf w
$$
where $\mathbf w$ inlucdes $w_0$.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
train = np.loadtxt('Data/hw8_train.txt')
test = np.loadtxt('Data/h8_test.txt')
N_train = train.shape[0] # number of examples in train set
N_test = test.shape[0] # number of examples in test set
print('The dimensions of the train data sample is :',train.shape)
print('The dimensions of the test data sample is :',test.shape)
```

```python
X = train[:,1:]
y = train[:,0]
Xt = test[:,1:]
yt = test[:,0]
```

```python
def one_vs_all(y,d):
    '''
    Inputs:
        y - input target vector
        d - digit
    Output:
        yn - (1) for all target equal to d and (-1) otherwise
    '''
    yn = np.zeros(len(y))
    
    for k in range(len(y)):
        if (y[k] == d):
            yn[k] = 1
        else:
            yn[k] = -1

    return yn
```

```python
def one_vs_one(X,y,d1,d2):
    '''
    Inputs:
        y - input target vector
        d1 - first digit
        d2 - second digit
    Output:
        yn - (1) for all target equal to d1; 
             (-1) for all target equal to d1; 
             (0) otherwise
    '''
    yn = np.zeros(len(y))
    
    for k in range(len(y)):
        if (y[k] == d1):
            yn[k] = 1
        else:
            if (y[k] == d2):
                yn[k] = -1
            else:
                yn[k] = 0
    
    XY = np.column_stack([yn,X])
    XY = XY[XY[:,0] != 0,:]
    X = XY[:,1:]
    y = XY[:,0]
    
    return X,y
```

```python
def x_transform(X):
    x1 = X[:,0]
    x2 = X[:,1]
    N = x1.shape[0]
    return np.column_stack([np.ones(N),
                            X,
                            X**2,
                            x1*x2])
```

```python
def error(X,y,w):        
    '''
    Binary classification error
    '''
    # Compute out-of-sample error
    N = X.shape[0]
    gt = np.where(np.dot(X,w) >= 0, 1, -1)
    error = gt - y
    E = np.count_nonzero(error)/N
       
    return E
```

```python
def lr_reg(X,y,lam):
    # Compute the weights
    m = X.shape[1]
    reg_term = lam*np.eye(m)
    w = np.dot(np.dot(np.linalg.inv(np.dot(X.T,X) + reg_term),X.T),y)
           
    return w
```

### Question 7
Set $\lambda = 1$ and do not apply a feature transform (i.e., use $z = x = (1, x_1, x_2)$). Which among the following classifiers has the lowest $E_{in}$?

```python
lam = 1 #lambda
digit = []
Ein = []
Eout = []

for i in range(10): # run for each digit vs all data set
    digit.append(i)
    d = i
    yd = one_vs_all(y,d)
    ytd = one_vs_all(yt,d)
    lr_w = lr_reg(X,yd,lam)
    Ein.append(error(X,yd,lr_w))
    Eout.append(error(Xt,ytd,lr_w))
```

```python
dt = {'digit':digit, 'Ein':Ein, 'Eout': Eout}
df = pd.DataFrame(data = dt)
df
```

## Radial Basis Functions
We experiment with the RBF model, both in regular form (Lloyd + pseudo-inverse) with $K$ centers:

$$
sign\bigg( \sum^{K}_{k=1} w_kexp(-\gamma|| \mathbf x-mu_k||^2) + b \bigg) 
$$

(notice that there is a bias term), and in kernel form (using the RBF kernel in hardmargin
SVM):

$$
sign\bigg( \sum_{\alpha_n > 0} \alpha_ny_n exp(-\gamma|| \mathbf x-x_n||^2) + b \bigg) 
$$

The input space is $\mathcal X = [-1,1] \times [-1,1]$ with uniform probability distribution, and the target is

$$
f(x)=sign(x_2-x_1+0.25sin(\pi x_1))
$$

which is slightly nonlinear in the $\mathcal X$ space. In each run, generate 100 training points at random using this target, and apply both forms of RBF to these training points.
Here are some guidelines:  
- Repeat the experiment for as many runs as needed to get the answer to be stable
(statistically away from flipping to the closest competing answer).  
- In case a data set is not separable in the $\mathcal Z$ space by the RBF kernel using hardmargin SVM, discard the run but keep track of how often this happens, if ever.  
- When you use Lloyds algorithm, initialize the centers to random points in $\mathcal X$ and
iterate until there is no change from iteration to iteration. If a cluster becomes empty, discard the run and repeat.

```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import random
from sklearn.svm import SVC # SVC implementation is based on libsvm
from numpy import linalg as la
```

Define function to generate random data using the target function $f$ given in the problem.

```python
f = lambda X: np.sign(X[:,1]-X[:,0]+0.25*np.sin(np.pi*X[:,0]))
```

```python
def gen_data(N,f):        
    while True:
        X = np.random.uniform(-1.,1., size=(N, 2))
        Y = f(X)
        if Y.min() != Y.max(): # ensure two class target
            break
            
    return X,Y
```

To do the problem we need to implement the RBF model in regular form and in kernel form. For the regular form we use Lloyd algorithm (L16s11), for which we first need to define a function which will iteratively find centers and then compute the $\Phi$ and finally $w$ according to the formulas given on L16s11. The steps are:  
1. Initialize random centers
2. Define which cluster each point belongs to
3. Compute new mean of each cluster
4. Re-run the cluster with new mean until there is no change in clusters for all data points  

Note that we only compute the run if there are no non-empty clusters.

```python
def my_kmeans(X,K):
    # Initialize random centeroid
    centers = np.random.uniform(-1.,1., size=(K, 2))
    
    S_old = np.zeros((K,1))
    while True:
        # label points to classes
        S = np.argmin(np.sqrt(np.sum((X-centers[:,np.newaxis])**2,
                                     axis=2)), 
                      axis = 0)
        
        # Check if all clusters are present else we get error 
        if len(np.unique(S)) == K:
            # Compute new centeroid for each k class
            centers = np.array([X[S==k].mean(axis=0) for k in range(K)])
        
            # Stop run if there is no change in labels and all labels are non-empty
            #if (S_old == S) and min(np.bincount(S))>0:
        if (S_old == S).all() :
            break
           
        S_old = S
    
    return S, centers
```

Define the $\Phi$ matrix.

```python
def phi_mat(X,centers,gamma):
    return np.exp(-gamma*la.norm(X[:, None] - centers, axis = 2)**2)
```

Define the errors for both the general and kernel based RBF.

```python
def rbf_L_E(X,y,Xt,yt,K,gamma):
    
    N = X.shape[0]
    Nt = Xt.shape[0]

    S, centers = my_kmeans(X,K)
    phi = np.column_stack([np.ones((X.shape[0],1)), phi_mat(X,centers,gamma)])
    w = la.lstsq(phi,y)[0]

    h = np.dot(phi,w)
    g = np.where(h >= 0, 1, -1)
    
    phit = np.column_stack([np.ones((Xt.shape[0],1)), phi_mat(Xt,centers,gamma)])
    ht = np.dot(phit,w)
    gt = np.where(ht >= 0, 1, -1)
    
    # Compute the errors
    Ein = np.count_nonzero(g-y)/N
    Eout = np.count_nonzero(gt-yt)/Nt

    return Ein, Eout
```

### Question 13
For $\gamma = 1.5$, how often do you get a data set that is not separable by the RBF kernel (using hard-margin SVM)? Hint: Run the hard-margin SVM, then check if the solution has $E_{in} \neq 0$.

```python
N = 100
gamma = 1.5
no_trials = 1000

cnt_E = []

for i in range(no_trials):
    
    X,y = gen_data(N,f)
    N = X.shape[0]
        
    m = SVC(kernel = 'rbf', gamma = gamma, coef0 = 1, C=np.inf)
    m = m.fit(X,y)
    g = m.predict(X)
    Ein = np.count_nonzero(g - y)/N
    if Ein > 0:
        cnt_E += 1
        
print(cnt_E)
```

<span style="color:blue">The answer is (a).</span>

### Question 14
If we use $K = 9$ for regular RBF and take $\gamma = 1.5$, how often does the kernel form beat the regular form (excluding runs mentioned in Problem 13 and runs with empty clusters, if any) in terms of $E_{out}$?

```python
K = 9
gamma = 1.5
N = 100
Nt = 1000
no_trials = 1000

Eout_L = []; Eout_K = []
cnt_E = 0
for i in range(no_trials):
    
    X,y = gen_data(N,f)
    N = X.shape[0]
    Xt,yt = gen_data(Nt,f)
    Nt = Xt.shape[0]
    
    # Kernel form
    m = SVC(kernel = 'rbf', gamma = gamma, C=np.inf)
    m = m.fit(X,y)
    gt = m.predict(Xt)
    Eoutk = np.count_nonzero(gt - yt)/Nt
    Eout_K.append(Eoutk)
    
    # Regular form
    EinL, EoutL = rbf_L_E(X,y,Xt,yt,K,gamma)
    Eout_L.append(EoutL)
    
    if Eoutk < EoutL:
        cnt_E += 1
        
print('The kernel form beats the regular form ',(cnt_E/no_trials)*100,'%')
```

```python
plt.scatter(range(no_trials),Eout_L, c='b', alpha = 0.4, edgecolor='k', label = 'Regular')
plt.scatter(range(no_trials),Eout_K, c='r',  alpha = 0.4, edgecolor='k', label = 'Kernel')
plt.legend()
plt.show()
```

<span style="color:blue">The answer is (e).</span>

## Bayesian Priors
### Question 19
Let $f \in [0,1]$ be the unknown probability of getting a heart attack for people in a certain population. Notice that $f$ is just a constant, not a function, for simplicity. We want to model $f$ using a hypothesis $h \in [0,1]$. Before we see any data, we assume that $P(h = f)$ is uniform over $h \in [0,1]$ (the prior). We pick one person from the population, and it turns out that he or she had a heart attack. Which of the following is true about the posterior probability that $h = f$ given this sample point?

**Answer**: By lecture we have the following definition for the posterior:

$$
P(h=f|\mathcal D)=\frac{P(\mathcal D|h=f)P(h=f)}{P(\mathcal D)}
$$  

We need to compute $P(h=f|\mathcal D)$. By the problem we have that $P(h=f)$ is $U(0,1)$ hence its density if equal to 1. To compute $P(\mathcal D|h=f)$ we use the fact that the event of having a heart attach is either *yes* or *no* i.e. it is a Bernoulli random variable with the probability of that event (having a heart attach) happening if $p$ and $(1-p)$ for no event. So $P(1|h=f) = h^k(1-k)^{1-k} = h$ since $k=1$. Now to compute $P(\mathcal D)$ we have the following:
$$
P(\mathcal D) = \int^{1}_{0}h(1)dh = \frac{h^2}{2}|^1_0=\frac{1}{2}
$$
Finally we have the posterior as follows:
$$
P(h=f|\mathcal D)=\frac{1}{2}h(1)=\frac{h}{2}
$$

<span style="color:blue">Answer is (b).</span>
